# Visualizing Probability Mass Function (PMF) for Variable 'Country'

```python
# Importing necessary libraries for plotting and numerical operations
import matplotlib.pyplot as plt
import numpy as np

# Given data: countries and their associated probabilities
countries = ['Italy', 'Australia', 'China', 'Spain', 'New Zealand']
probabilities = [0.18, 0.25, 0.38, 0.10, 0.09]

# Create a figure and a set of subplots (bar chart and pie chart)
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Bar Chart: Plotting the Probability Mass Function (PMF)
axs[0].bar(countries, probabilities, color='green')
axs[0].set_xlabel('Countries')  # Label for the X-axis
axs[0].set_ylabel('Probability')  # Label for the Y-axis
axs[0].set_title('Probability Mass Function for Variable Country')  # Chart title
axs[0].set_xticks(range(len(countries)))  # Positioning the tick marks on the X-axis
axs[0].set_xticklabels(countries, rotation=45)  # Setting the labels for the tick marks
axs[0].set_ylim([0, 0.5])  # Setting the range for the Y-axis

# Annotating each bar with the corresponding probability value
for i, prob in enumerate(probabilities):
    axs[0].text(i, prob-0.01, str(prob), ha='center')

# Pie Chart: Plotting the Probability Mass Function (PMF)
axs[1].pie(probabilities, labels=countries, autopct='%1.2f%%', startangle=90)
axs[1].set_title('Pie Chart for Variable Country')  # Chart title
axs[1].axis('equal')  # Ensures the pie chart is a circle

# Show both the bar chart and the pie chart
plt.show()

```
