```python
# Importing the required libraries
import matplotlib.pyplot as plt
import numpy as np

# Data: Country proportions
countries = ['Italy', 'Australia', 'China', 'Spain', 'New Zealand']
probabilities = [0.18, 0.25, 0.38, 0.10, 0.09]

# Plotting the bar chart with countries on the x-axis and probabilities on the y-axis
plt.figure(figsize=(10, 6))
plt.bar(countries, probabilities, color='green')

plt.xlabel('Countries')
plt.ylabel('Probability')
plt.title('Categorical Distribution of Students by Country')
plt.ylim([0, max(probabilities) + 0.1])  # Setting the y-axis limit to make the chart more readable

# Annotate the bars with the actual probabilities
for i, prob in enumerate(probabilities):
    plt.text(i, prob - 0.01, str(prob), ha='center')

plt.show()
```
