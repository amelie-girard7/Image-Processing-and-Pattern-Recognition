# Visualizing Probability Mass Function (PMF) for Variable 'Country'

```python
# Importing necessary libraries for plotting and numerical operations
import matplotlib.pyplot as plt
import numpy as np

# Given data: countries and their associated probabilities
countries = ['Italy', 'Australia', 'China', 'Spain', 'New Zealand']
probabilities = [0.18, 0.25, 0.38, 0.10, 0.09]

# Create a figure and a single subplot (bar chart)
fig, axs = plt.subplots(1, 1, figsize=(8, 6))  # Removed the second subplot

# Bar Chart: Plotting the Probability Mass Function (PMF)
axs.bar(countries, probabilities, color='green')
axs.set_xlabel('Countries')  # Label for the X-axis
axs.set_ylabel('Probability')  # Label for the Y-axis
axs.set_title('Probability Mass Function for Variable Country')  # Chart title
axs.set_xticks(range(len(countries)))  # Positioning the tick marks on the X-axis
axs.set_xticklabels(countries, rotation=45)  # Setting the labels for the tick marks
axs.set_ylim([0, 0.5])  # Setting the range for the Y-axis

# Annotating each bar with the corresponding probability value
for i, prob in enumerate(probabilities):
    axs.text(i, prob-0.01, str(prob), ha='center')

# Show the bar chart
plt.show()
```
