```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the Gaussian PDFs and CDFs
params = [
    {'mean': -2, 'variance': 0.5, 'color': 'blue'},
    {'mean': 0, 'variance': 1, 'color': 'red'},
    {'mean': 1, 'variance': 1.5, 'color': 'green'}
]

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(18, 12))

# Generate x values
x = np.linspace(-5, 5, 1000)

# Plot Gaussian PDFs and CDFs
for param in params:
    mean = param['mean']
    variance = param['variance']
    color = param['color']
    
    # Plot PDF
    y_pdf = norm.pdf(x, mean, np.sqrt(variance))
    axs[0].plot(x, y_pdf, color=color, label=f"={mean}, ={variance}")
    
    # Plot CDF
    y_cdf = norm.cdf(x, mean, np.sqrt(variance))
    axs[1].plot(x, y_cdf, color=color, label=f"={mean}, ={variance}")

# PDF Plot settings
axs[0].set_title('Gaussian Probability Density Functions')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Probability Density')
axs[0].legend()

# CDF Plot settings
axs[1].set_title('Gaussian Cumulative Density Functions')
axs[1].set_xlabel('x')
axs[1].set_ylabel('Cumulative Density')
axs[1].legend()

# Show plots
plt.tight_layout()
plt.show()
```
