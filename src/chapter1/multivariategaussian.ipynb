```python
import numpy as np
import matplotlib.pyplot as plt

# Define mean and covariance matrix
mu = np.array([0, 0])
Sigma = np.array([[0.61, 0.48], [0.48, 0.64]])

# Generate 10,000 samples
samples = np.random.multivariate_normal(mu, Sigma, 10000)

# Scatter plot
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.2, color='blue')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Scatter Plot of 10,000 2D Gaussian Samples')
plt.grid(True)

# Show plot
plt.show()
```
